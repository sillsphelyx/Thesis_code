```python
# Packages
%pylab inline
%load_ext memory_profiler
from pomegranate import BayesianNetwork
import seaborn, time
seaborn.set_style('whitegrid')
from pomegranate import *
import numpy
import pandas as pd
import pygraphviz as pgv
import seaborn as sns
```

```python
# Data 
df = pd.read_csv('total_leak_pred_data.csv')
df.head()
```

```python
df1 = pd.read_csv('meter_leak_pred_data.csv')
df1.head()
```

```python
#Fitting BN to the categorised data setAvg_Temp_disc
model = BayesianNetwork.from_samples(df , state_names = [ 'Temp_range' ,'Prs_us_disc','Total_leak'], algorithm='exact')
```

```python
#model = BayesianNetwork.from_samples(X, algorithm='exact')
print(model.structure)
#model.plot()

plt.figure(figsize=(14, 10))
model.plot()
plt.show()
```
