# SMOTE  for class balancing

```python
import numpy as np
from collections import Counter
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification
```

```python
# Importing data and applying  SMOTE (synthetic minority over-sampling technique )

# for reproducibility purposes
seed = 100
# SMOTE number of neighbors
k = 1

df = pd.read_csv('meter_leak_pred_data_lagged.csv', encoding='utf-8', engine='python')
# make a new df made of all the columns, except the target class
X = df.loc[:, df.columns != 'meterleak_count']
y = df.meterleak_count
#sm = SMOTE(random_state=42, ratio = 1.0)
sm = SMOTE(sampling_strategy='auto', k_neighbors=k, random_state=seed)
X_res, y_res = sm.fit_resample(X, y)
```

```python
# creating a data frame with the over-sampled data and writting to a csv file 
df = pd.concat([pd.DataFrame(X_res), pd.DataFrame(y_res)], axis=1)
# rename the columns
df.columns = ['Prs_ds_disc', 'Prs_ds_disc_t1', 'Prs_us_disc', 'Prs_us_disc_t1',
        'MinTemp_disc', 'MinTemp_disc_t', 'MaxTemp_disc', 'MaxTemp_disc_t1',
        'Avg_Temp_disc', 'Avg_Temp_disc_t1', 'Temp_range_disc',
        'Temp_range_disc_t', 'Rain_disc', 'Rain_disc_t', 'meterleak_count_t','meterleak_count']
df.to_csv('SMOTE_meterleak.csv', index=False, encoding='utf-8')
```
